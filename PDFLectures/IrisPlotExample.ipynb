```python
import numpy as np
from matplotlib import pyplot as plt
```

Note that it is important to first import the necessary libraries.

```python
def readIris(path):
    iris = np.loadtxt(path,delimiter=',',dtype='str')
    features = iris[:,:-1]
    labels = iris[:,-1]
    return features,labels
```

```python
# adapted from: https://scipy-lectures.org/packages/scikit-learn/auto_examples/plot_iris_scatter.html
path = "./iris.data"
features,labels = readIris(path)
features = features.astype(float)
u, indices = np.unique(labels, return_inverse=True)
feature_names = ['sepal length', 'sepal width', 'petal length', 'petal width']
# The indices of the features that we are plotting
x_index = 0 
x_name = feature_names[x_index]
target_names = np.array(['Iris-setosa','Iris-versicolor','Iris-virginica'])
y_index = 1 
y_name = feature_names[y_index]


# this formatter will label the colorbar with the correct target names
formatter = plt.FuncFormatter(lambda i, *args: target_names[int(i)])

plt.figure(figsize=(5, 4))
plt.scatter(features[:, x_index], features[:, y_index], c=indices)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(x_name)
plt.ylabel(y_name)

plt.tight_layout()
plt.show()
```
